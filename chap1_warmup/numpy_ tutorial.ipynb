#### 2.建立一个一维数组 a 初始化为[4,5,6], (1)输出a 的类型（type）(2)输出a的各维度的大小（shape）(3)输出 a的第一个元素（值为4）

```python
b = np.array([[4, 5, 6], [1, 2, 3]])
print(b.shape)
print(b[0, 0])
print(b[0, 1])
print(b[1, 1])
```

```python
print(x * y)
print(np.multiply(x, y))
print(np.dot(x, y))

# 非方阵示例
x_ns = np.array([[1, 2, 3], [4, 5, 6]], dtype=np.float64)
y_ns = np.array([[7, 8, 9], [10, 11, 12]], dtype=np.float64)
print(np.multiply(x_ns, y_ns))   # 对应元素乘
print(np.dot(x_ns, y_ns.T))      # 矩阵乘
```

#### 24,画图，y=x*x 其中 x = np.arange(0, 100, 0.1) （提示这里用到  matplotlib.pyplot 库）

```python
import matplotlib.pyplot as plt
x_plot = np.arange(0, 100, 0.1)
y_plot = x_plot * x_plot
plt.plot(x_plot, y_plot)
plt.title("y = x^2")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

#### 25.画图。画正弦函数和余弦函数， x = np.arange(0, 3 * np.pi, 0.1)(提示：这里用到 np.sin() np.cos() 函数和 matplotlib.pyplot 库)

```python
x_trig = np.arange(0, 3 * np.pi, 0.1)
plt.plot(x_trig, np.sin(x_trig), label="sin(x)")
plt.plot(x_trig, np.cos(x_trig), label="cos(x)")
plt.title("sin(x) and cos(x)")
plt.xlabel("x")
plt.ylabel("value")
plt.legend()
plt.grid(True)
plt.show()
```
